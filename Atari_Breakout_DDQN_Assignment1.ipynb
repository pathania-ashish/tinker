```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import gym
import matplotlib.pyplot as plt

gamma = 0.99            # Gamma Discount Factor, importance of future rewards in the agent's decision-making process
epsilon = 1.0           # Probability of taking a random action instead of the best-known action
epsilon_min = 0.1       # Agent will never take a completely random action once epsilon has reached this value
epsilon_max = 1.0       # Value of epsilon will start at this value and gradually decrease over time
epsilon_interval = (
    epsilon_max - epsilon_min
)                       # Rate at which to reduce chance of random action being taken
batch_size = 32         # The number of experiences that are sampled from the experience buffer and used to update the agent's policy during training
max_steps_per_episode = 10000
                        # Maximum number of time steps that the agent is allowed to take in each episode of the game

env = gym.make('Breakout-v4',render_mode='rgb_array')
                       ## From Open AI Gym library imported a 'Breakout-v4' environment which when rendered returns a RGB Array  
num_actions = env.action_space.n
                       ## number of possible actions in the game ...how 4?
```

```python
color = np.array([210, 164, 74]).mean()

## (Yellowish-orange hue) It can be used as a threshold to identify pixels in the game that correspond to the color of interest
## Represents a grayscale value that is used to indicate certain features of the game screen that the agent needs to pay attention to or ignore we will decide

def preprocess_state(state):

    image = state[1:176:2, ::2]     ##  it takes every other row and column starting from the second row , subset image od size (88, 80, 3)

    #convert the image to greyscale
    image = image.mean(axis=2)     ## axis=2 specifies that the mean should be calculated along the third axis, which represents the color channels.
    
    #improve image contrast
    image[image==color] = 0        ## removes the pixels in image that match the color value.
                                   ## remove some of the noise in the image that is not relevant for the agent's task

    #normalize the image
    image = (image - 128) / 128 - 1  ## pixel values (0-256) normalized to be in the range [-1, 1]

    return image
```

```python
num_actions = 4


def create_q_model():
    # Network defined by the Deepmind paper
    inputs = layers.Input(shape=(88, 80, 1,))           

    # Convolutions on the frames on the screen
    ## The model applies a series of convolutional layers to the input tensor to extract features from the frames
    ## The model has three convolutional layers with 32, 64, and 64 filters, respectively, and kernel sizes of 8x8, 4x4, and 3x3, respectively

    layer1 = layers.Conv2D(32, 8, strides=4, activation="relu")(inputs)
    layer2 = layers.Conv2D(64, 4, strides=2, activation="relu")(layer1)
    layer3 = layers.Conv2D(64, 3, strides=1, activation="relu")(layer2)


    layer4 = layers.Flatten()(layer3)

    layer5 = layers.Dense(512, activation="relu")(layer4)
    action = layers.Dense(num_actions, activation="linear")(layer5) 
    ## ## a linear activation function, which allows the model to output a Q-value for each possible action.

    return keras.Model(inputs=inputs, outputs=action)


model = create_q_model() # main network, q value networks

model_target = create_q_model() # target network

## Remember:  The target network is updated less frequently than the main network, and its weights are copied from the main network periodically. 
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.00025, clipnorm=1.0)


action_history = []           # A list to store the actions taken by the agent during the current episode.
state_history = []            # A list to store the preprocessed game frames that correspond to the states visited by the agent during the current episode.
state_next_history = []       # A list to store the preprocessed game frames that correspond to the next states visited by the agent during the current episode
rewards_history = []          # A list to store the rewards received by the agent during the current episode.
done_history = []             # A list to store the boolean values that indicate whether the episode has ended.
episode_reward_history = []   # A list to store the total rewards obtained by the agent in each episode.
running_reward = 0            # A running average of the total rewards obtained by the agent over all episodes.
episode_count = 0             # A counter for the number of episodes completed.
frame_count = 0               # A counter for the total number of frames processed so far during training

epsilon_greedy_frames = 1000000.0     # The number of frames over which to anneal the exploration rate from epsilon to epsilon_min.
 
update_target_network = 10000  
# The number of frames after which to update the weights of the target Q-network with those of the primary Q-network.

# Using huber loss for stability
loss_function = keras.losses.Huber()


while True:  
    env.reset()                                 # resets the game environment to its initial state.
    state = preprocess_state(env.render())      # gets the current state and and preprocessing it 
    episode_reward = 0
    

    for timestep in range(max_steps_per_episode):       # To iterates over the steps of each episode
        
        frame_count += 1                                # This variable is used later to track when to update the target network.

        if epsilon > np.random.rand(1)[0]:
            
            action = np.random.choice(num_actions)

            ## If current value of the epsilon parameter is greater than
            #  randomly generated value between 0 and 1  it takes a random action. 
            #  Otherwise, the agent selects the action with the highest Q-value according to the current model

        else:
            
            state_tensor = tf.convert_to_tensor(state)
            state_tensor = tf.expand_dims(state_tensor, 0)
            action_probs = model(state_tensor, training=False)

            ## Here The model is used to predict the Q-values for the current state
            # Since training=False, the model's weights are not updated during prediction.

            action = tf.argmax(action_probs[0]).numpy()
            # Action with the highest Q-value is selected
            
        
        epsilon -= epsilon_interval / epsilon_greedy_frames
        epsilon = max(epsilon, epsilon_min)
        # epsilon states The 'Exploration rate' 
        # The probability of the agent choosing a random action over the action suggested by the policy network
        # It's value is decreasing as agent gains more experience and policy netwrok become more accurate.


        state_next, reward, done, other = env.step(action) # env.step returns the next state, reward, done flag and other info

        state_next = preprocess_state(state_next)

        episode_reward += reward

        
        action_history.append(action)
        state_history.append(state)
        state_next_history.append(state_next)
        done_history.append(done)
        rewards_history.append(reward)
        state = state_next

        
        if len(done_history) > batch_size:
          # checks if there are enough samples in the replay buffer to form a batch of size batch_size.
            
            
            indices = np.random.choice(range(len(done_history)), size=batch_size)
            # randomly selects batch_size indices from the list of indices 
            
            state_sample = np.array([state_history[i] for i in indices])
            #contains the state information for each sample in the batch.

            state_next_sample = np.array([state_next_history[i] for i in indices])
            # contains the next state information for each sample in the batch.

            rewards_sample = [rewards_history[i] for i in indices]
            # creates a list of length batch_size that contains the reward for each sample in the batch

            action_sample = [action_history[i] for i in indices]
            #  creates a list of length batch_size that contains the action taken for each sample in the batch.

            done_sample = tf.convert_to_tensor(
                [float(done_history[i]) for i in indices]
            )
            # creates a tensor of shape (batch_size,) that contains a boolean value .
            # indicating whether each sample in the batch is terminal (1.0) or not (0.0)
            
            ## DDQN Modifications


            # # Use the current model to select actions for the next state
            # # and the target model to estimate the Q value for that action
           
            
            future_rewards_main = model.predict(state_next_sample, verbose=0)
            # It uses the main Q-network to predict the Q-values of the next state (state_next_sample) for all possible actions

            q_values_next_target = model_target.predict(state_next_sample, verbose=0)
            # This line uses the target Q-network to predict the Q-values of the next state for all possible actions.

            best_actions = tf.argmax(future_rewards_main, axis=1)
            #  This line finds the index of the action that has the highest predicted Q-value for each sample in the batch

            target_q_values = tf.gather_nd(q_values_next_target, tf.stack((tf.range(batch_size), tf.cast(best_actions, tf.int32)), axis=1))
            # This line uses the indices of the best actions found above to extract the Q-values of those actions from the target Q-network.
            

            updated_q_values = rewards_sample + gamma * target_q_values
         
            updated_q_values = updated_q_values * (1 - done_sample) - done_sample

            masks = tf.one_hot(action_sample, num_actions)
            ## Creates a one-hot encoding of the actions taken in the current state
            #  action_sample contains the actions taken in the current state
            #  num_actions is the number of possible actions in the environment
            # one-hot encoding is used to select the Q-value of the action taken in the current state.
            

            with tf.GradientTape() as tape:
                
                q_values = model(state_sample)
                # to obtain the predicted Q-values for each action in the current state.

                q_action = tf.reduce_sum(tf.multiply(q_values, masks), axis=1)
                # calculates the Q-value of the chosen action based on the predicted Q-values of all actions.

                loss = loss_function(updated_q_values, q_action)
                # calculates the loss between the updated Q-values and the predicted Q-values for the chosen actions.


            
            grads = tape.gradient(loss, model.trainable_variables)
            # computes the gradients of the loss with respect to the trainable variables of the model.

            optimizer.apply_gradients(zip(grads, model.trainable_variables))
            # updates the trainable variables of the model using the gradients computed above and the optimizer.

        if frame_count % update_target_network == 0:    # checks if it is time to update the target network based   

            model_target.set_weights(model.get_weights()) # updates the weights of the target network to be the same as the main model.
            
            template = "running reward: {:.2f} at episode {}, frame count {}"
            print(template.format(running_reward, episode_count, frame_count))


        if done: # checks if the episode has ended
            break
            
    episode_reward_history.append(episode_reward)
    running_reward = np.mean(episode_reward_history)
    

    episode_count += 1
    
    

    if episode_count > 50:
        print("Solved at episode {}!".format(episode_count))
        break
```

```python
import matplotlib.pyplot as plt

# Plot episode rewards
plt.plot(episode_reward_history)
plt.title("Episode Rewards")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```

```python
# Plot running average reward
running_avg_reward = [np.mean(episode_reward_history[max(0, i-10):i+1]) for i in range(len(episode_reward_history))]
plt.plot(running_avg_reward)
plt.title("Running Average Reward (Over 10 Episodes)")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```

```python
# Plot frame count vs. running average reward
frame_count = [i*max_steps_per_episode for i in range(1, len(running_avg_reward)+1)]
plt.plot(frame_count, running_avg_reward)
plt.title("Running Average Reward vs. Frame Count")
plt.xlabel("Frame Count")
plt.ylabel("Reward")
plt.show()
```
